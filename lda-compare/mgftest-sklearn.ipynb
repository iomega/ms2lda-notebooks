# Convert corpus into sparse matrix

```python
import json
```

```python
lda_dict = json.load(open('METABOLOMICS-SNETS-e9eaf6ec-download_clustered_spectra-main.lda.json'))
corpus = json.load(open('METABOLOMICS-SNETS-e9eaf6ec-download_clustered_spectra-main.corpus.json'))
```

```python
from scipy.sparse import lil_matrix
import numpy as np
```

```python
m = lil_matrix((len(corpus), len(lda_dict['word_index'])))
```

```python
for doc, words in corpus.items():
    for word, score in words.items():
        m[lda_dict['doc_index'][doc], lda_dict['word_index'][word]] = score
```

```python
K = 300
n_its = 1000
tf = m.tocsr()
```

```python
from sklearn.decomposition import LatentDirichletAllocation
```

```python
%time lda = LatentDirichletAllocation(n_components=K , max_iter=n_its, learning_method='online', learning_offset=50., random_state=0).fit(tf)
```
