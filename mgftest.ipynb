```python
from ms2lda_feature_extraction import MakeBinnedFeatures, LoadMGF
from lda import VariationalLDA
```

```python
loader = LoadMGF(min_ms1_intensity = 0.0,
                        min_ms2_intensity = 5000,
                        mz_tol=5,
                        rt_tol=10,
                        peaklist=None)
```

```python
ms2_file = 'METABOLOMICS-SNETS-e9eaf6ec-download_clustered_spectra-main.mgf'
K = 300
n_its = 1000
```

```python
ms1, ms2, metadata = loader.load_spectra([ms2_file])
```

```python
fm = MakeBinnedFeatures()
corpus, word_mz_range = fm.make_features(ms2)
```

```python
word_mz_range
```

```python
corpus = corpus[corpus.keys()[0]]
```

```python
vlda = VariationalLDA(corpus=corpus, K=K, normalise=1000.0)
vlda.run_vb(n_its=n_its, initialise=True)

lda_dict = vlda.make_dictionary(metadata=metadata, features=word_mz_range)
```
