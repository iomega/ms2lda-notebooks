```python
import json
```

```python
# Generated by gensim in mgftest-gensim.ipynb 
lda_dict = json.load(open('METABOLOMICS-SNETS-e9eaf6ec-download_clustered_spectra-main.jensim.lda.json'))
```

```python
from basicviz.models import Experiment, User, BVFeatureSet, UserExperiment, JobLog
```

```python
user = User.objects.get(username='stefanv')
```

```python
featureset = BVFeatureSet.objects.first()
```

```python
new_experiment = Experiment(name='gensim3', description='inserted from nb', status='0', experiment_type='0',
                                           experiment_ms2_format='2', featureset=featureset, ms2_file='METABOLOMICS-SNETS-e9eaf6ec-download_clustered_spectra-main.mgf'
                                          )
```

```python
new_experiment.save()
```

```python
UserExperiment.objects.create(user=user, experiment=new_experiment, permission='edit')
```

```python
JobLog.objects.create(user = user, experiment = new_experiment, tasktype = 'Uploaded and ' + new_experiment.experiment_type)
```

```python
from load_dict_functions import load_dict
```

```python
load_dict(lda_dict, new_experiment, feature_set_name=featureset.name)
```

```python
new_experiment.status = '1'
new_experiment.save()
```
